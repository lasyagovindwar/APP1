```python
#Understand the dataset structure.
import pandas as pd

# Step 1: Load the dataset (replace 'yourfile.csv' with your actual file)
df = pd.read_csv("/content/Data_set 2 - Copy.csv")

# Step 2: Display basic information
print(df.info())

# Optional: Look at the first few rows
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Step 1: Extract gender column
gender_counts = df['gender'].value_counts()

# Step 2: Plot bar chart
gender_counts.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('gender')
plt.ylabel('Count')
plt.show()

# Optional Pie Chart
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Step 1 & 2: Identify numerical columns and calculate stats
numerical_columns = df.select_dtypes(include=['int64', 'float64'])
print(numerical_columns.describe())
```

```python
#Find the most common investment type.
# Step 1 & 2: Frequency analysis
investment_counts = df['Investment_Avenues'].value_counts() # Changed column name to 'Investment_Avenues'
print(investment_counts)

# Optional Visualization
investment_counts.plot(kind='bar', color='orange')
plt.title('Most Preferred Investment_Avenue')
plt.xlabel('Avenue')
plt.ylabel('Number of Participants')
plt.show()
```

```python
# Combine all reasons into a single list
reasons = pd.concat([
    df['Reason_Equity'],
    df['Reason_Mutual'],
    df['Reason_Bonds'],
    df['Reason_FD']
])
```

```python
# Clean: lowercase and strip spaces
reasons_clean = reasons.astype(str).str.lower().str.strip()
```

```python
# Count occurrences of each reason
reason_counts = reasons_clean.value_counts()
print(reason_counts)
```

```python
import matplotlib.pyplot as plt

reason_counts.head(10).plot(kind='bar', title='Top Reasons for Investment', color='skyblue')
plt.xlabel('Reason')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Clean and analyze savings objectives
savings_objectives = df['What are your savings objectives?'].astype(str).str.lower().str.strip()

# Step 2: Count frequency
objective_counts = savings_objectives.value_counts()

# Step 3: Display result
print(objective_counts)
```

```python
import matplotlib.pyplot as plt

objective_counts.head(10).plot(kind='bar', title='Top Savings Objectives', color='lightgreen')
plt.xlabel('Savings Objective')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Task 7: Common Information Sources
sources = df['Source'].astype(str).str.lower().str.strip()
source_counts = sources.value_counts()
print("Common Information Sources:\n", source_counts)

# Optional: Visualization
source_counts.plot(kind='bar', title='Investment Information Sources', color='purple')
plt.xlabel('Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#Task 9: Expectations from Investments
expectations = df['Expect'].astype(str).str.lower().str.strip()
expectation_counts = expectations.value_counts()
print("Expectations from Investments:\n", expectation_counts)

# Optional: Visual
expectation_counts.plot(kind='bar', title='Investment Return Expectations', color='teal')
plt.xlabel('Expectation')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
duration_map = {
    'less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'more than 5 years': 6
}

# Convert to lowercase first to match mapping keys
df['Duration'] = df['Duration'].str.lower().str.strip()
df['Duration_Num'] = df['Duration'].map(duration_map)
```

```python
correlation_data = df[['age', 'Duration_Num']]
correlation_matrix = correlation_data.corr()
print("Correlation Matrix:\n", correlation_matrix)

# Optional: Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
